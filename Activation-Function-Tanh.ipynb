```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def tanhActivation(number):
    data = []
    for i in number:
        res = (math.exp(i) - math.exp(-i)) / (math.exp(i) + math.exp(-i))
        data.append(res)
    return data
```

```python
def plot_tanh(number=np.arange(-10.0, 10.0, .5)):
    data_plot = tanhActivation(number)
    plt.plot(number, data_plot)
    
    ax = plt.gca()
    plt.show()
```

```python
plot_tanh(np.arange(-25.0, 10.0, 0.5))
plot_tanh()
```
